## Product Details
1. Product Id
2. Name of product
3. Price of product
4. Quantity

```python
# Importing to note the time.
import time

# Opening and reading the file.
fd = open('Inventory.txt', 'r')
products = fd.read().split('\n')
fd.close()

# Taking user inputs and user details
usr_name = input("Enter your name: ")
usr_phone = input("Enter your phone no: ")
usr_mail = input("Enter your mail: ")
usr_prod_input = input("Enter product id: ")
usr_quant = input("Enter the quantity: ")

# Creating an empty product list
updated_prod_lst = []

# Going through each product details
for product in products:

    prod = product.split(',')
    
    # Checking if product exists or not
    if (prod[0] == usr_prod_input):
        
        # if we have enough quantity
        if(int(usr_quant) <= int(prod[3])):
            
            print("----------------------------------")
            print("Product Name      : ", prod[1])
            print("Price             : ", prod[2])
            print("Quantity          : ", usr_quant)
            print("----------------------------------")
            print("Billing Amount    : ", int(prod[2]) * int(usr_quant))
            print("----------------------------------")
            
            # Updating Inventory 
            prod[3] = str(int(prod[3]) - int(usr_quant))
            
            # Adding sales details in Sales.txt file
            fd = open('Sales.txt', 'a')
            
            sales_details = usr_name+","+usr_phone+","+usr_mail+","+prod[1]+","+usr_prod_input+","+usr_quant+","+str(int(prod[2]) * int(usr_quant))+","+time.ctime()+"\n"
            
            fd.write(sales_details)
            fd.close()
            
        # If we dont have enough quantity
        else:
            print("Sorry, we're not having enough quantity.")
            print("We're having only",prod[3],"quantity.")
            print("Would you like to purchase it?")
            
            ch = input("Press Y/N : ")
            
            # If user want to purchase the remaining quantity
            if(ch == 'Y' or ch == 'y'):
                print("----------------------------------")
                print("Product Name      : ", prod[1])
                print("Price             : ", prod[2])
                print("Quantity          : ", prod[3])
                print("----------------------------------")
                print("Billing Amount    : ", int(prod[2]) * int(prod[3]))
                print("----------------------------------")
                
                
                # Adding sales details in Sales.txt file
                fd = open('Sales.txt', 'a')

                sales_details = usr_name+","+usr_phone+","+usr_mail+","+prod[1]+","+usr_prod_input+","+prod[3]+","+str(int(prod[2]) * int(prod[3]))+","+time.ctime()+"\n"

                fd.write(sales_details)
                fd.close()
                
                # Updating Inventory
                prod[3] = '0'
                
            else:
                print("Thanks")
    
    # Updating Inventory List
    updated_prod_lst.append(prod)
    
lst = []

# Converting inventory list into string format
for i in updated_prod_lst:
    pr = i[0]+','+i[1]+','+i[2]+','+i[3]+'\n'
    lst.append(pr)

# Removing the last \n from the list
lst[-1] = lst[-1][:-1]

# Updating Inventory.txt file
fd = open('Inventory.txt', 'w')

for i in lst:
    fd.write(i)
    
fd.close()

print("----------------------------------")
print("Inventory Updated")
```
